line plot

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
z = [5,6,7,9,3]

plt.plot(x,y, linewidth = 2.0, linestyle = ':', color='r', alpha= .5, marker = 'o' )
plt.plot(x,z, linewidth = 2.0, linestyle = ':', color='b', alpha= .5, marker = 'o' )
plt.title('line plot demo')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.legend(['line1', 'line2'], loc = 'best')
plt.show()
```

```python
import numpy as np
a = np.arange(0,5,1)

x = [1,2,3,4,5]
y = [2,4,6,8,10]
z = [5,6,7,9,3]

# plot graph 1
plt.subplot(1,3,1)
plt.plot(a,x, linewidth = 2.0, linestyle = ':', color='r', alpha= .5, marker = 'o' )
plt.title('graph-1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.legend(['line1'], loc = 'best')

# plot graph 2
plt.subplot(1,3,2)
plt.plot(a,y, linewidth = 2.0, linestyle = ':', color='g', alpha= .5, marker = 'o' )
plt.title('graph-2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.legend(['line2'], loc = 'best')

# plot graph 3
plt.subplot(1,3,3)
plt.plot(a,z, linewidth = 2.0, linestyle = ':', color='b', alpha= .5, marker = 'o' )
plt.title('graph-3')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.legend(['line3'], loc = 'best')
```

bar plot

```python
data = {'apples':20, 'mangoes':15, 'lemon':30, 'oranges':10}

names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize=(8,5))
plt.bar(names[:2], values[:2], color = 'orange')
plt.bar(names[2:], values[2:], color = 'skyblue')
# plt.bar(names[2:], values[2:], color = 'skyblue')
plt.show()
```

scatter plot

```python
a = [10,20,30,40,50,60,70,80]
x = [5,3,2,5,6,1,4,2]
y = [1,2,3,4,3,2,1,2]

plt.scatter(a,x)
plt.scatter(a,y, c='seagreen', s = 100, alpha=1, marker='4')
plt.show()
```

histogram

```python
x = [10,12,16,19,11,20,26,28,30,38,35,34,45,60,68,64,62,70,78,75,79,85,94,95]
plt.hist(x, bins=[0,20,40,60,80,100], color='gold')
plt.show()
```

pie chart

```python
label = ['Dog','Cat','Wolf','Lion']
sizes = [50,45,60,80]

plt.pie(sizes, labels=label, autopct='%1.2f%%', shadow=True, startangle= 90, explode=(0,0,0,0.2))
plt.title('pie chart demo')
plt.show()
```
